Importing the modules

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
print('Modules are imported.')
```

Importing covid-19 dataset

```python
covid_dataset_csv=pd.read_csv("D:\Python\Covid-19 data analysis\covid19_Confirmed_dataset.csv")
covid_dataset_csv.head()
```

To check for shape of the dataframe


```python
covid_dataset_csv.shape
```

Delete unwanted colums

```python
df=covid_dataset_csv.drop(["Lat","Long"],axis=1,inplace=True)
covid_dataset_csv.head()
```

```python
covid_dataset_aggrigated=covid_dataset_csv.groupby("Country/Region").sum()
covid_dataset_aggrigated.head()
```

```python
covid_dataset_aggrigated.shape
```

Visualizing data related to country

```python

covid_dataset_aggrigated.loc["India"].plot()

covid_dataset_aggrigated.loc["Italy"].plot()

plt.legend()
```

```python
covid_dataset_aggrigated.loc["India"].diff().plot()
```

Finding max inection rate

```python
covid_dataset_aggrigated.loc["India"].diff().max()
```

Finding max infection rate of all the countries 

```python
countries= list(covid_dataset_aggrigated.index)
max_infection_rate=[]
for i in countries:
    max_infection_rate.append(covid_dataset_aggrigated.loc[i].diff().max())
max_infection_rate
```

```python
covid_dataset_aggrigated["max_infection_rate"]=max_infection_rate
covid_dataset_aggrigated.head()
```

```python
covid_data = pd.DataFrame(covid_dataset_aggrigated["max_infection_rate"])
covid_data.head()
```

Importing the WorldHappinessReport.csv dataset


```python
happiness_report_csv= pd.read_csv("D:\Python\Covid-19 data analysis\worldwide_happiness_report.csv")
happiness_report_csv.head
```

```python
happiness_report_csv.drop(useless_col,axis=1,inplace=True)
happiness_report_csv.head()
```
